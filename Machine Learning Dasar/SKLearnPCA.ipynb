```python
#MEMBAGI DATASET
from sklearn import datasets
from sklearn.model_selection import train_test_split
iris = datasets.load_iris()
atribut =iris.data
label = iris.target

#bagi dataset menjadi train set dan tese set
x_train, x_test, y_train, y_test = train_test_split(atribut, label, test_size=0.2, random_state=1)
```

```python
#LATIH MODEL TANDAP PCA

from sklearn import tree

decision_tree = tree.DecisionTreeClassifier()
model_pertama = decision_tree.fit(x_train, y_train)
model_pertama.score(x_test, y_test)
```

```python
#LATIH MODEL DENGAN PCA

from sklearn.decomposition import PCA

#membuat object PCA dengan 4 principal component
pca =PCA(n_components=4)

#mengaplikasikan PCA pada dataset
pca_atributes = pca.fit_transform(x_train)

#melihat variance dari setiap atribut
pca.explained_variance_ratio_
```

```python
# PCA dengan 2 principal component
pca = PCA(n_components = 2)
x_train_pca = pca.fit_transform(x_train)
x_test_pca = pca.fit_transform(x_test)
```

```python
#EVALUASI HASIL KEDUA MODEL
model2 = decision_tree.fit(x_train_pca, y_train)
model2.score(x_test_pca, y_test)
```
